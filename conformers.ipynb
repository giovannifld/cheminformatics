```python
from rdkit import Chem
from rdkit.Chem import AllChem

def generate_and_compare_conformers(smiles_string, num_conformers=50, max_attempts=1000):
    mol = Chem.MolFromSmiles(smiles_string)
    if mol is None:
        print("Invalid SMILES string.")
        return

    mol = Chem.AddHs(mol) # Add hydrogens for more accurate conformer generation

    # Generate conformers
    AllChem.EmbedMultipleConfs(mol, numConfs=num_conformers, maxAttempts=max_attempts)

    # Optimize conformers and store energies
    energies = []
    for i, conf in enumerate(mol.GetConformers()):
        # Perform MMFF energy minimization
        ff = AllChem.MMFFGetMoleculeForceField(mol, AllChem.MMFFGetMoleculeProperties(mol), confId=conf.GetId())
        if ff:
            ff.Minimize()
            energy = ff.CalcEnergy()
            energies.append((i, energy))
        else:
            print(f"Could not set up force field for conformer {i}")
            energies.append((i, float('inf')))

    # Sort conformers by energy
    energies.sort(key=lambda x: x[1])

    print(f"Conformer energies for {smiles_string}:")
    for i, energy in energies:
        print(f"  Conformer {i}: {energy:.2f} kcal/mol")

    if energies:
        print(f"\nLowest energy conformer: Conformer {energies[0][0]} with energy {energies[0][1]:.2f} kcal/mol")
    else:
        print("No conformers were generated or optimized.")

# Example usage:
smiles = "CCO"
# smiles = "CC(=O)Oc1ccccc1C(=O)O" # Aspirin
generate_and_compare_conformers(smiles)
```
