```python
import rdkit
from rdkit import Chem
from rdkit.Chem.Draw import IPythonConsole
from IPython.display import display

# Example SMILES string
smiles_string = "CCO"

# Convert SMILES to a molecule object
molecule = Chem.MolFromSmiles(smiles_string)

# Display the molecule structure
if molecule:
    print(f"Structure for SMILES: {smiles_string}")
    display(molecule)
else:
    print(f"Could not generate structure for SMILES: {smiles_string}")


# You can change the smiles_string variable to visualize other molecules.
# For example:
# smiles_string = "c1ccccc1"
# molecule = Chem.MolFromSmiles(smiles_string)
# if molecule:
#     display(molecule)
```
